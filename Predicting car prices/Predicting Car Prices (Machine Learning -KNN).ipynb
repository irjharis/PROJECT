```python
features = ['wheel-base', 'length', 'width', 'curb-weight', 'bore', 'horsepower',
       'city-mpg', 'highway-mpg']

target = ["price"]
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knn = KNeighborsRegressor(n_neighbors=5) # k is by default 5

knn.fit(train[features], train[target])  # training 
prediction = knn.predict(test[features]) # prediction
```

```python
test["p_price"] = prediction
```

```python
test["absolute_error"] = abs(test["price"] - test["p_price"])
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse = mean_squared_error(test["price"],prediction)
rmse = np.sqrt(mse)
rmse
```

```python
k_rmse = {}

for k in range(3,20): # grid search- hyperparamer optimatization
    knn = KNeighborsRegressor(n_neighbors=k) 
    knn.fit(train[features], train[target])  # training 
    prediction = knn.predict(test[features]) # prediction
    mse = mean_squared_error(test["price"],prediction)
    rmse = np.sqrt(mse)
    k_rmse[k] = rmse
```

```python
k_rmse
```

```python
min(k_rmse,key= k_rmse.get )
```

```python
k_r = {}

for k in range(3,20): # grid search- hyperparamter optimatization
    knn = KNeighborsRegressor(n_neighbors=k,algorithm='brute',) 
    knn.fit(train[features], train[target])  # training 
    prediction = knn.predict(test[features]) # prediction
    mse = mean_squared_error(test["price"],prediction)
    rmse = np.sqrt(mse)
    k_r[k] = rmse
```

```python
# metric='minkowski'


k_rm = {}

for k in range(3,20): # grid search- hyperparamter optimatization
    knn = KNeighborsRegressor(n_neighbors=k,algorithm='brute',metric='euclidean') 
    knn.fit(train[features], train[target])  # training 
    prediction = knn.predict(test[features]) # prediction
    mse = mean_squared_error(test["price"],prediction)
    rmse = np.sqrt(mse)
    k_rm[k] = rmse
```
